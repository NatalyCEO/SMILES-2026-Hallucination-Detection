# SMILES: запуск из архива

1. **Runtime → Change runtime type → GPU** (T4).
2. Загрузите zip проекта в `/content/` (иконка папки слева → Upload).
3. В следующей ячейке укажите **`ARCHIVE_NAME`** — имя вашего файла.
4. Запускайте ячейки по порядку.

```python
# Имя загруженного zip в /content/ (поменяйте на своё)
ARCHIVE_NAME = "SMILES-2026-Hallucination-Detection.zip"

import os
import zipfile
from pathlib import Path

ARCHIVE_PATH = Path("/content") / ARCHIVE_NAME
if not ARCHIVE_PATH.is_file():
    raise FileNotFoundError(
        f"Нет {ARCHIVE_PATH}. Загрузите архив в /content/ или поправьте ARCHIVE_NAME."
    )

extract_root = Path("/content") / "_smiles_extract"
if extract_root.exists():
    import shutil

    shutil.rmtree(extract_root)
extract_root.mkdir(parents=True)

with zipfile.ZipFile(ARCHIVE_PATH, "r") as z:
    z.extractall(extract_root)

hits = list(extract_root.rglob("solution.py"))
if not hits:
    raise FileNotFoundError(
        "В архиве не найден solution.py. Проверьте, что zip — корень проекта."
    )

PROJECT_ROOT = hits[0].parent.resolve()
os.chdir(PROJECT_ROOT)
print("Рабочая папка:", PROJECT_ROOT)
assert (PROJECT_ROOT / "data" / "dataset.csv").is_file(), "Нет data/dataset.csv"
assert (PROJECT_ROOT / "data" / "test.csv").is_file(), "Нет data/test.csv"
print("OK: данные на месте.")
```
